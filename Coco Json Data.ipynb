# Converting csv file 'train_annot.csv','val_annot.csv' to json files  

CoCo Json format was originally specified by CoCo Object Detection Competition. It stores object detection data in json format. The json contains following four fields.For training object detection model under Detectron2, only images, categories and annotations fields are mandatory , info and licenses field can be ignored.

{
"info": info,
"images": [image],
"annotations": [annotation],
"licenses": [license],
"categories": [category]
}


Images field contains list of dicts, with one dict for each image. It assigns image id to image name which is used in annotations field. Its structure for one single image is as follows:
{
"id": int,
"width": int,
"height": int,
"file_name": str,
"license": int,
"flickr_url": str,
"coco_url": str,
"date_captured": datetime,
}
For training on detectron2, only id, width, height and file_name field are mandatory.

Categories field contains list of dicts, with one dict for each category. It assigns category id to category name which is used in annotations field. Its structure for one single category is as follows:
{
"id": int,
"name": str,
"supercategory": str,
}


Annotations field contains list of dicts, with one dict for each label. We have to provide unique annotation id for each annotations. And we have to provide associated image id and category id. For detection task we have to provide bounding box coordinates and its area.in case of object detection task segmentation field can be left empty.
annotation{
"id": int,
"image_id": int,
"category_id": int,
"segmentation": [], 
"area": float,
"bbox": [x,y,width,height],
"iscrowd": 0,
}
So I convert  tensorflow object detection csv data into Coco Json format using following code.

```python
#creating json file for 'train_annot.csv'
```

```python
data=pd.read_csv(r"E:\data\data (1)\data\train_annot.csv")
```

```python
data
```

```python
data= data.rename(columns={'x-axis': 'x', 'y-axis': 'y'})
```

```python
save_json_file = 'train_annot_new1.json'
```

```python
images = []
categories = []
annotations = []
```

```python
#category = {}
#category = {}
#category["supercategory"] = 'none'
#category["id"] = 0
#category["name"] = 'None'
#categories.append(category)
```

```python
data['fileid'] = data['name'].astype('category').cat.codes
data['categoryid']= pd.Categorical(data['class'],ordered= True).codes
data['categoryid'] = data['categoryid']+1
data['annid'] = data.index
```

```python
def image(row):
    image = {}
    image["height"] = row.image_height
    image["width"] = row.image_width
    image["id"] = row.fileid
    image["file_name"] = row.name
    return image
```

```python
def category(row):
    category = {}
    category["supercategory"] = 'None'
    category["id"] = row.categoryid
    category["name"] = row[1]
    return category
```

```python
def annotation(row):
    annotation = {}
    area = None
    annotation["segmentation"] = []
    annotation["iscrowd"] = 0
    annotation["area"] = area
    annotation["image_id"] = row.fileid

    annotation["bbox"] = [row.x, row.y, row.width,row.height ]

    annotation["category_id"] = row.categoryid
    annotation["id"] = row.annid
    return annotation
```

```python
for row in data.itertuples():
    annotations.append(annotation(row))
```

```python
imagedf = data.drop_duplicates(subset=['fileid']).sort_values(by='fileid')
for row in imagedf.itertuples():
    images.append(image(row))
```

```python
catdf = data.drop_duplicates(subset=['categoryid']).sort_values(by='categoryid')
for row in catdf.itertuples():
    categories.append(category(row))
```

```python
data_coco = {}
data_coco["images"] = images
data_coco["categories"] = categories
data_coco["annotations"] = annotations
```

```python
json.dump(data_coco, open(save_json_file, "w"), indent=4)
```
